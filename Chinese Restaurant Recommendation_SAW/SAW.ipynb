```python
import pandas as pd
import numpy as np

df = pd.read_csv('akb_fix.csv')
a = df.shape
print (a)
df = df.drop_duplicates()
df.shape
```

```python
# Step 1: Ambil kolom yang akan digunakan
df_values = df[['Rating', 'Cost', 'Delivery time(mins)']].copy()

# Step 2: Transformasi kriteria cost sesuai Persamaan (2)
df_values['Cost'] = df_values['Cost'].min() / df_values['Cost']
df_values['Delivery time(mins)'] = df_values['Delivery time(mins)'].min() / df_values['Delivery time(mins)']

# Step 3: Noemalisasi kriteria sesuai Persamaan (1)
df_values['Rating'] = df_values['Rating'] / df_values['Rating'].max()
df_values['Cost'] = df_values['Cost'] / df_values['Cost'].max()
df_values['Delivery time(mins)'] = df_values['Delivery time(mins)'] / df_values['Delivery time(mins)'].max()

# Step 4: Hitung proporsi
prop = df_values / df_values.sum(axis=0)

# Step 5: Hitung Entropi
k = 1 / np.log(len(df_values)) 
entropy = -k * (prop * np.log(prop + 1e-12)).sum(axis=0)  

# Step 6: Hitung derajat dispersi dan bobot
d = 1 - entropy
weights = d / d.sum()

print("Nilai Entropi untuk masing-masing kriteria:")
print(entropy)

print("\nBobot kriteria berdasarkan Metode Entropi:")
print(weights)
```

```python
# Step 7: Hitung nilai terbobot untuk setiap alternatif
df_saw = df_values.copy()  # hasil normalisasi
df_saw['Rating'] = df_saw['Rating'] * weights['Rating']
df_saw['Cost'] = df_saw['Cost'] * weights['Cost']
df_saw['Delivery time(mins)'] = df_saw['Delivery time(mins)'] * weights['Delivery time(mins)']

# Step 8: Hitung skor akhir (penjumlahan semua kriteria)
df['Final Score'] = df_saw.sum(axis=1)

# Step 9: Urutkan berdasarkan skor tertinggi
df_sorted = df.sort_values(by='Final Score', ascending=False).reset_index(drop=True)

print("\nSkor akhir berdasarkan metode SAW:")
df_sorted[['Name', 'Rating', 'Cost', 'Delivery time(mins)', 'Final Score']]
```
